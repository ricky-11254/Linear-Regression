```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
hello = pd.read_csv('/content/AccidentSeverity.csv')
x = hello[['casualty_severity']]
y = hello['age_of_casualty']
regressor = LinearRegression()
regressor.fit(x, y)
hello.plot(x='casualty_severity', y= 'age_of_casualty', style ='o', label='DataPoints')
plt.xlabel ='casualty_severity'
plt.ylabel = 'age_of_casualty'
plt.plot(x, regressor.predict(x), color ='red', linewidth = 2,label = 'Regression Line')
plt.legend()
plt.show()
```
